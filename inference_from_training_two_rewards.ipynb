```
# 12 is cumulative sum of time_stamps_survived - [1,2,3.....] it is total_t
k = np.ones(df_1.shape[0])
df_1[12] = df_1[1]-k


df_1[12] = df_1[12].cumsum()
# df[12].head()

# 13 is total episodic reward =  survival_status * time_stams_survived * avg_vel_in_episode
# df[13] = df[5] * df[1]* df[2]

##  13 total episodic reward =   time_stams_survived * avg_episode_reward
df_1[13] = df_1[3] * df_1[1]

# df_1[13].head()


# 12 is cumulative sum of time_stamps_survived - [1,2,3.....] it is total_t
k = np.ones(df_2.shape[0])
df_2[12] = df_2[1]-k


df_2[12] = df_2[12].cumsum()
# df[12].head()

# 13 is total episodic reward =  survival_status * time_stams_survived * avg_vel_in_episode
# df[13] = df[5] * df[1]* df[2]

##  13 total episodic reward =   time_stams_survived * avg_episode_reward
df_2[13] = df_2[3] * df_2[1]

# df_1[13].head()







```

### TOTAL EPISODIC REWARD




```
alpha = 0.02
mvar_1 = np.zeros((df_1.shape[0]),dtype=np.float)
for i in range(df_1.shape[0]):
   if i == 0:
     mvar_1[i] = df_1.loc[i,13]
   else:
     mvar_1[i] = alpha * df_1.loc[i,13] + (1.0 - alpha) * mvar_1[i-1] 

# plt.plot(df_1.loc[:,0], df_1.loc[:,13])
# plt.plot(df_1.loc[:,0], mvar)
# plt.xlabel('episode_number', fontsize=20)
# plt.ylabel('episode reward', fontsize=20)

alpha = 0.02
mvar_2 = np.zeros((df_2.shape[0]),dtype=np.float)
for i in range(df_2.shape[0]):
   if i == 0:
     mvar_2[i] = df_2.loc[i,13]
   else:
     mvar_2[i] = alpha * df_2.loc[i,13] + (1.0 - alpha) * mvar_2[i-1] 




```

```
fig, ax = plt.subplots()
ax.plot(df_1.loc[:,0], mvar_1 , label = "Gaussian reward",color = '#ff7f0e')
ax.plot(df_2.loc[:,0], mvar_2 , label = "Exponential rise and fall reward", color ='violet')
ax.set_xlabel('episode_number', fontsize=20)
ax.set_ylabel('episodic reward ', fontsize=20)
leg = ax.legend()
```
